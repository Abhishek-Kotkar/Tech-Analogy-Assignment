## Data Generation

```python
import pandas as pd
import numpy as np
import random

# Define the dataset size
num_records = 10000

# Generate the data
data = {
    "Lead ID": ["LD" + str(i + 1) for i in range(num_records)],
    "Location": np.random.choice(
        ["Ahmedabad", "Surat", "Mumbai", "Delhi", "Bangalore", "Chennai", "Pune"], num_records
    ),
    "College": np.random.choice(
        ["Manipal University", "My_Equation Institute", "SVM College", "Unique Academy", "Pune University"], num_records
    ),
    "Year of Study": np.random.choice([1, 2, 3, 4], num_records),
    "Program Interest": np.random.choice(
        ["Data Science", "Robotics", "AI", "Electric Vehicle", "Cybersecurity"], num_records
    ),
    "Lead Source": np.random.choice(
        ["Instagram", "LinkedIn", "College Collaboration", "Google Form", "Mass-Mailing", "Whatsapp"], num_records
    ),
}

# Create the DataFrame
df = pd.DataFrame(data)

# Save to a CSV file (optional)
df.to_csv("generated_leads.csv", index=False)

print("Dataset generated successfully!")
```

## Data PreProcessing

- Basically this data does not have any anomalies and null values.
- first rule :- you have to check the null values in the dataset.
- second rule :- if null values are less than 5% in the dataset you have to drop that rows ny using (df.dropna()).
- third rule :- if null values are greater than 5% in the dataset, you have to check column wise null values.
- fourt rule :- if null values are greater than 40% in any column , you have to drop that column.
- fifth rule :- if null values are less than 40% in any column you have to replace null values by (mean,mode,median).
- you have to check duplicate records in the data.
